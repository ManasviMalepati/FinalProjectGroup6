```python
import pandas as pd
import os

##Requires - data directory FOLDER: Census D
data_dir = 'Data/Census Data'

file_years = [
    ('2015_CensusData.csv', 2015),
    ('2016_CensusData.csv', 2016),
    ('2017_CensusData.csv', 2017),
    ('2018_CensusData.csv', 2018),
    ('2019_CensusData.csv', 2019),
    ('2021_CensusData.csv', 2021),
    ('2022_CensusData.csv', 2022)
]

df_list = []

for file_name, year in file_years:
    df = pd.read_csv(os.path.join(data_dir, file_name))
    df['Year'] = year
    df_list.append(df)

census_data_torename = pd.concat(df_list, ignore_index=True)
census_data = census_data_torename.drop(columns = ['B19051_001M' , 'B19051_002M', 'B19051_003M', 'Unnamed: 8'])

census_data.columns = [
    'census_geo_id', 'census_geo_name', 'average_household_earnings', 'avg_household_earnings_with_earnings', 'avg_household_earnings_no_earnings', 'year'
]

census_data = census_data[1:]
```
